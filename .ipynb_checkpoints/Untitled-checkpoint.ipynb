### Math Scores by Grade
__(using pivot table)__

```python
# creating a pivot table calculating mean of math_score for index as grade and column as school name
df_math_grade = df_stu.pivot_table( values = 'math_score', index='grade',  columns='school_name', aggfunc='mean').T
```

```python

df_math_grade
```

```python
# since 9th comes in the end in lexical analysis of grade names, we need to shift back column 9th in the begining
col_sequence = df_math_grade.columns.tolist()
col_sequence = col_sequence[-1:] + col_sequence[:-1]
col_sequence
```

```python
df_math_grade = df_math_grade[col_sequence]
```

```python
#reformatting all columns to 2 digits after decimal 
type(df_math_grade)
decimal_format = df_math_grade.columns.tolist() # all columns need to be reformatted

for item in decimal_format:
    df_math_grade[item] = df_math_grade[item].map('{:,.2f}'.format)
    
df_math_grade
```
