```python
import pandas as pd
from pathlib import Path
```

```python
data_folder=Path("")
data_file1 = "04-Pandas_Instructions_PyCitySchools_Resources_schools_complete.csv"
data_file2 = "04-Pandas_Instructions_PyCitySchools_Resources_students_complete.csv"

file1 = data_folder / data_file1
file2 = data_folder / data_file2
```

```python
df_sch = pd.read_csv(file1)
df_stu = pd.read_csv(file2)
```

### District Summary

```python
index_list = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score',
                'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']
```

```python
tot_school = df_stu['school_name'].nunique()
tot_student = df_stu['Student ID'].count()  # take note that students may have same names across grades/schools
tot_budget = df_sch['budget'].sum()
avg_math_sco = df_stu['math_score'].mean()
avg_read_sco = df_stu['reading_score'].mean()
tot_pass_math = df_stu.loc[df_stu['math_score']>=70, :].count()['Student ID']
tot_pass_read = df_stu.loc[df_stu['reading_score']>=70, :].count()['Student ID']
#tot_pass_overall = df_stu.loc[(df_stu['math_score']>=70) & (df_stu['reading_score']>=70) , :].count()['Student ID']
tot_pass_overall = (tot_pass_math+tot_pass_read)/2
```

```python
dict1 = {'Total Schools':tot_school, 'Total Students':tot_student, 'Total Budget':tot_budget, 'Average Math Score':avg_math_sco, 
                     'Average Reading Score':avg_read_sco, '% Passing Math': 100*tot_pass_math/tot_student, 
                     '% Passing Reading': 100*tot_pass_read/tot_student, 
                     '% Overall Passing Rate':100*(tot_pass_overall/tot_student)}
dict1
```

```python
df_dist_summary = pd.DataFrame(dict1, index=[0])
df_dist_summary['Total Budget'] = df_dist_summary['Total Budget'].map('${:,.2f}'.format)
df_dist_summary['Average Math Score'] = df_dist_summary['Average Math Score'].map('{:,.2f}'.format)
df_dist_summary['Average Reading Score'] = df_dist_summary['Average Reading Score'].map('{:,.2f}'.format)
df_dist_summary['% Passing Math'] = df_dist_summary['% Passing Math'].map('{:,.2f}'.format)
df_dist_summary['% Passing Reading'] = df_dist_summary['% Passing Reading'].map('{:,.2f}'.format)
df_dist_summary['% Overall Passing Rate'] = df_dist_summary['% Overall Passing Rate'].map('{:,.2f}'.format)
df_dist_summary
```

### School Summary

```python
df_sch.head()
```

```python
df_sch = df_sch.set_index('school_name')
```

```python

df_sch.columns
```

```python
# new_columns = ['Per student Budget','Average Math Score', 'Average Reading Score', '% Passing math', '% Passing Reading', 'Overall Passing Rate']
list_school = df_sch.index.tolist()

for school in list_school:
    df_sch.loc[school, 'Average Math Score'] = df_stu.loc[df_stu['school_name']==school, 'math_score'].mean()
    df_sch.loc[school, 'Average Reading Score'] = df_stu.loc[df_stu['school_name']==school, 'reading_score'].mean()
    df_sch.loc[school, '% Passing Math'] = 100 * ( df_stu.loc[(df_stu['school_name']==school) & (df_stu['math_score']>=70) , :].count()['Student ID'])/df_sch.loc[school, 'size']
    df_sch.loc[school, '% Passing Reading'] = 100 * ( df_stu.loc[(df_stu['school_name']==school) & (df_stu['reading_score']>=70) , :].count()['Student ID'])/df_sch.loc[school, 'size']
    #df_sch.loc[school, '% Overall Passing Rate'] = 100 * ( df_stu.loc[(df_stu['school_name']==school) & (df_stu['math_score']>=70) & (df_stu['reading_score']>=70) , :].count()['Student ID'])/df_sch.loc[school, 'size']
    df_sch.loc[school, '% Overall Passing Rate'] = ( df_sch.loc[school,'% Passing Math']+df_sch.loc[school,'% Passing Reading'])/2
# better way to define for large data sets as below:
# dollar_format = ['budget']
# decimal_format = ['Average Math Score', 'Average Reading Score','% Passing Math', '% Passing Reading', '% Overall Passing Rate']
# for item in dollar_format:
#     df_sch[item] = df_sch[item].map('${:,.2f}'.format)
# for item in decimal_format:
#     df_sch[item] = df_sch[item].map('{:,.2f}'.format)

```

### Math Scores by Grade
__(using pivot table)__

```python
# creating a pivot table calculating mean of math_score for index as grade and column as school name
df_math_grade = df_stu.pivot_table( values = 'math_score', index='grade',  columns='school_name', aggfunc='mean').T
```

```python

df_math_grade.head()
```

```python
# since 9th comes in the end in lexical analysis of grade names, we need to shift back column 9th in the begining
col_sequence = df_math_grade.columns.tolist()
new_col_sequence = col_sequence[-1:] + col_sequence[:-1]
new_col_sequence
```

```python
df_math_grade = df_math_grade[new_col_sequence]
```

```python
#reformatting all columns to 2 digits after decimal 
type(df_math_grade)
decimal_format = df_math_grade.columns.tolist() # all columns need to be reformatted

for item in decimal_format:
    df_math_grade[item] = df_math_grade[item].map('{:,.2f}'.format)
    
df_math_grade
```

## Summary: Top Observable Trends

1. Contrary to popular belief, Schools with lower per student spending are performing better !!
2. Schools of small & medium sizes (<2000 students) are performing better in comparison to bigger size schools.
3. Charter schools are performing much better in comparison to Distric schools. Poor Math score performance in District school is a major  reason for this trend.
